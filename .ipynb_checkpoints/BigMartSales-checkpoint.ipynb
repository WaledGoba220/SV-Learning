## Data Preprocessing (Handle Missing Value)

```python
data['Item_Weight'].mean()
```

```python
data['Item_Weight'].fillna(data['Item_Weight'].mean(), inplace=True)
```

```python
data['Outlet_Size'].mode()
```

```python
mode_of_Outlet_size = data.pivot_table(values='Outlet_Size', columns='Outlet_Type', aggfunc=(lambda x: x.mode()[0]))
```

```python
mode_of_Outlet_size
```

```python
miss_values = data['Outlet_Size'].isnull()   
```

```python
miss_values
```

```python
data.loc[miss_values, 'Outlet_Size'] = data.loc[miss_values,'Outlet_Type'].apply(lambda x: mode_of_Outlet_size[x])
```

## Data Preprocessing (Label Encoder)

```python
le = LabelEncoder()
```

```python
## ==> Fatma Solution
## a = ['Item_Identifier','Item_Fat_Content','Item_Type','Outlet_Identifier','Outlet_Size','Outlet_Location_Type','Outlet_Type']
## for cols in a:
   # print(cols)
## data[cols] =  le.fit_transform(data[cols])
```

```python
data['Item_Identifier'] = le.fit_transform(data['Item_Identifier'])

data['Item_Fat_Content'] = le.fit_transform(data['Item_Fat_Content'])

data['Item_Type'] = le.fit_transform(data['Item_Type'])

data['Outlet_Identifier'] = le.fit_transform(data['Outlet_Identifier'])

data['Outlet_Size'] = le.fit_transform(data['Outlet_Size'])

data['Outlet_Location_Type'] = le.fit_transform(data['Outlet_Location_Type'])

data['Outlet_Type'] = le.fit_transform(data['Outlet_Type'])
```

```python
data.head(10)
```

## Splitting Features and Target (Inputs Vs Output)

```python
X = data.drop(columns='Item_Outlet_Sales', axis=1)
Y = data['Item_Outlet_Sales']
```

```python
X
```

```python
Y
```

## Splitting Training Data and Testing Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

```python
print(Y.shape)
print(Y_train.shape)
print(Y_test.shape)
```

## Building Model

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,Y_train)
```

## Make Prediction 

```python
## ==> prediction on Training data
training_data_prediction = lr.predict(X_train)
```

```python
## ==>  R squared Error
error_score = metrics.r2_score(Y_train, training_data_prediction)
print("R squared Error : ", error_score)
```

```python
plt.scatter(Y_train, training_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title(" Actual Prices vs Predicted Prices")
plt.show()
```

```python
## ==> prediction on Training data
test_data_prediction = lr.predict(X_test)
```

```python
## ==> R squared Error
error_score = metrics.r2_score(Y_test, test_data_prediction)
print("R squared Error : ", error_score)
```

```python
plt.scatter(Y_test, test_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title(" Actual Prices vs Predicted Prices")
plt.show()
```
